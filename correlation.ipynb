```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Exercicio 1
X = 'Idade'
Y = 'Peso'
df = pd.DataFrame([[71, 17], [65, 17], [70, 18], [57, 17], [66, 19]], columns=[Y, X])
df.head()
```

```python
# a) Construa o diagrama de dispersão
df.plot.scatter(x=X, y=Y)
```

```python
# b) Calcule o coeficiente de correlação (com 3 casas decimais). Qual a interpretação desse valor?
# O valor de R mostra que há uma correlação muito baixa entre as variáveis peso e altura
x = np.array(df[X])
y = np.array(df[Y])
r = np.corrcoef(x, y)[0][1]
print(r)
```

```python
# c) Escreva a equação de regressão linear do tipo y=Ax+B.
modelo = LinearRegression()
modelo.fit(df[[X]], df[[Y]])
print(modelo.coef_)
print(modelo.intercept_)

# Make predictions using the testing set
y_pred = modelo.predict(df[['Idade']])

# Plot outputs
plt.scatter(df[['Idade']], df[['Peso']])
plt.plot(df[['Idade']], y_pred)

plt.show()

# equação: y = modelo.coef_*x + modelo.intercept_
# equação: y = 40.5 + 1.4375x
```
